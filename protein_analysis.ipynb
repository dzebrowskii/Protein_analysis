```python
fig, axes = plt.subplots(3,3, figsize=(18,12))
axes = axes.flatten()

for ax, column in zip(axes, df.columns):
    
    counts, bin_edges = np.histogram(df[column], bins=bins, density=True)
    pdf = counts / sum(counts)
    cdf = np.cumsum(pdf)

    ax.plot(bin_edges[1:], pdf, label='PDF', color='blue')
    ax.plot(bin_edges[1:], cdf, label='CDF', color='green')

    ax.set_title(f'PDF / CDF {column}')
    ax.set_xlabel(column)  
    ax.set_ylabel('Probability / Cumulative Probability')
    ax.legend(loc='best')
    ax.grid(True)  
        

plt.tight_layout()  
plt.show()
```

```python
fig, axes = plt.subplots(3,2, figsize=(18,12))
axes = axes.flatten()

for ax, column in zip(axes, df_grouped_all.columns):
    
    counts, bin_edges = np.histogram(df_grouped_all[column], bins=bins, density=True)
    pdf = counts / sum(counts)
    cdf = np.cumsum(pdf)

    ax.plot(bin_edges[1:], pdf, label='PDF', color='blue')
    ax.plot(bin_edges[1:], cdf, label='CDF', color='green')

    ax.set_title(f'PDF / CDF {column}')
    ax.set_xlabel(column)  
    ax.set_ylabel('Probability / Cumulative Probability')
    ax.legend(loc='best')
    ax.grid(True)  
        

plt.tight_layout()  
plt.show()
```
