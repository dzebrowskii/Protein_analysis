## Wizualizacja danych

```python
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))
plt.title('Boxplot dla Spożycia Różnych Źródeł Protein')
plt.xlabel('Źródła Protein')
plt.ylabel('Spożycie (g/osoba)')
sns.boxplot(data = df)
```

```python
plt.figure(figsize=(12, 6))
plt.title('Boxplot dla Spożycia Różnych Źródeł Protein')
plt.xlabel('Źródła Protein')
plt.ylabel('Spożycie (g/osoba)')
sns.violinplot(data = df)
```

```python
bins = round(np.sqrt(df.shape[0]))
axes = df.hist(bins=bins, figsize=(12, 13), layout=(3, 3))
plt.gcf().suptitle('Rozkład spożycia różnych źródeł protein')
for ax in axes.flatten():
    ax.set_xlabel('Spożycie (g/osoba)')
    ax.set_ylabel('Liczba')

plt.show()
```

```python
fig, axes = plt.subplots(3,3, figsize=(18,12))
axes = axes.flatten()

for ax, column in zip(axes, df.columns):
    
    counts, bin_edges = np.histogram(df[column], bins=bins, density=True)
    pdf = counts / sum(counts)
    cdf = np.cumsum(pdf)

    ax.plot(bin_edges[1:], pdf, label='PDF', color='blue')
    ax.plot(bin_edges[1:], cdf, label='CDF', color='green')

    ax.set_title(f'PDF / CDF {column}')
    ax.set_xlabel(column)  
    ax.set_ylabel('Probability / Cumulative Probability')
    ax.legend(loc='best')
    ax.grid(True)  
        

plt.tight_layout()  
plt.show()
```
