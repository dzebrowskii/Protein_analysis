```python
fig, axes = plt.subplots(3,3, figsize=(18,12))
axes = axes.flatten()

for ax, column in zip(axes, df.columns):
    
    counts, bin_edges = np.histogram(df[column], bins=bins, density=True)
    pdf = counts / sum(counts)
    cdf = np.cumsum(pdf)

    ax.plot(bin_edges[1:], pdf, label='PDF', color='blue')
    ax.plot(bin_edges[1:], cdf, label='CDF', color='green')

    ax.set_title(f'PDF / CDF {column}')
    ax.set_xlabel(column)  
    ax.set_ylabel('Probability / Cumulative Probability')
    ax.legend(loc='best')
    ax.grid(True)  
        

plt.tight_layout()  
plt.show()
```

# PCA

## PCA dla covariancji 

```python
cova_matrix = df.cov()
cova_matrix
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=df.shape[1])
pca.fit(df)
eigenvalues = pca.explained_variance_
explained_variance_ratio = pca.explained_variance_ratio_
cumulative_variance = pca.explained_variance_ratio_.cumsum()

pca_results = pd.DataFrame({
    'Component': ['Comp{}'.format(i+1) for i in range(len(eigenvalues))],
    'Eigenvalue': eigenvalues,
    'Difference': [eigenvalues[i] - eigenvalues[i+1] if i < len(eigenvalues)-1 else '-' for i in range(len(eigenvalues))],
    'Proportion': explained_variance_ratio,
    'Cumulative': cumulative_variance
})
pca_results
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(eigenvalues) + 1), eigenvalues, 'o-', color='blue')
plt.title('Scree Plot of Eigenvalues after PCA')
plt.xlabel('Number of Components')
plt.ylabel('Eigenvalue')
plt.axhline(y=1, color='red', linestyle='--')

plt.show()
```

```python
loadings = pca.components_.T
explained_variance_by_variable = np.sum(loadings**2, axis=1)
unexplained_variance_by_variable = 1 - explained_variance_by_variable
variable_names = df.columns
pca_loadings_df = pd.DataFrame(loadings, columns=[f'Comp{i+1}' for i in range(9)], index=variable_names)
pca_loadings_df['Unexplained'] = unexplained_variance_by_variable
pca_loadings_df
```

```python
X_pca = pca.transform(df)
components_to_plot = [i for i, ev in enumerate(pca.explained_variance_) if ev > 1]

df_to_plot = pd.DataFrame(X_pca[:, components_to_plot], 
                          columns=[f'Component {i+1}' for i in components_to_plot])

sns.pairplot(df_to_plot)
plt.show()
```

```python
def count_passing_cars(road):
    # Inicjalizacja liczników
    cars_on_left = 0
    passing_cars = 0

    # Iteracja przez każdy element w wektorze
    for car in road:
        if car == 0:
            # Samochód jadący lewym pasem
            cars_on_left += 1
        elif car == 1:
            # Samochód jadący prawym pasem - dodaj liczbę samochodów na lewym pasie
            passing_cars += cars_on_left

    return passing_cars

# Przykładowe użycie
road = [0, 1, 1, 0, 1, 0, 0]
print(count_passing_cars(road))  
```
